```python
import pandas as pd
import numpy as np
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('Corrected_New_Features.csv',index_col='Sr_No')
```

```python
df.head()
```

```python
# Dropping the unnecessary columns
df.drop(['Unnamed: 0'],axis=1, inplace=True)
```

```python
df.head()
```

```python
# VADER sentiment analysis tool for getting pos, neg and neu.
def sentimental_Score(sentence):
    analyzer = SentimentIntensityAnalyzer()
    vs = analyzer.polarity_scores(sentence)
    score=vs['compound']
    if score >= 0.5:
        return 'pos'
    elif (score > -0.5) and (score < 0.5):
        return 'neu'
    elif score <= -0.5:
        return 'neg'
```

```python
df['Sentiment'] = df.Review_Text.apply(sentimental_Score)
```

```python
df.head()
```

```python
# Exporting the dataset containing sentiments to a csv file
df.to_csv('ReviewText_Sentiments.csv')
```
