## Modelling using XGBoost Regressor

```python
# Importing Necessary Modules
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

```python
# Importing the dataset containing unigram feature set
df = pd.read_csv('Unigram_Feature.csv',index_col='Sr_No')
```

### Getting an understanding of the data

```python
df.head()
```

```python
# Dropping the un-necessary columns
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

### Splitting into Train and Test data

```python
# All predictors in X
X=df.iloc[:,:-1]
# The target variables 'h' in y
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Splitting into train and test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=50,random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

### XGBoost Regressor: Modelling the train data

```python
# Training the XG Boost Regressor
xgb = XGBRegressor(n_estimators = 100, random_state = 0)
xgb.fit(X_train, y_train)
```

### Predicting on Test Data

```python
# Predicting on test data
y_pred = xgb.predict(X_test)
```

```python
y_pred
```

```python
# Creating a dataframe to compare predicted and test values
pd.DataFrame({"y_pred": y_pred, "y_test": y_test})
```

### Calculating Mean Absolute Percentage Error (MAPE)

```python
# Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```
