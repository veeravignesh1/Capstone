### Experimenting by removing Upvote and Downvote columns
Now we will experiment by removing Upvote and Downvote and then see the MAPE results.

```python
# Importing Necessary Modules
import pandas as pd
import numpy as np
import spacy
nlp=spacy.load('en_core_web_sm')
from sklearn.model_selection import train_test_split
from spacy.pipeline import SentenceSegmenter
import re
import warnings
warnings.filterwarnings('ignore')
import textblob
from textblob import TextBlob
from spacy.lang.en import English

# Customization of View
pd.options.display.max_columns = 999
```

```python
# Importing the dataset containing unigram feature set
df = pd.read_csv('Unigram_Feature.csv',index_col='Sr_No')
```

### Getting an understanding of the data

```python
df.head()
```

```python
# Dropping the un-necessary columns as well as Upvote and Downvote for our Experiment.
df.drop(['Unnamed: 0','Upvote','Downvote'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

### Splitting into Train and Test data

```python
# All predictors in X
X=df.iloc[:,:-1]
# The target variables 'h' in y
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Splitting into train and test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=50,random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

### Random Forest Regressor: Modelling the train data

```python
# Import the model
from sklearn.ensemble import RandomForestRegressor
```

```python
# Training the Random Forest Regressor
rf = RandomForestRegressor(n_estimators = 100, random_state = 0)
rf.fit(X_train, y_train)

# Getting the feature importance
feature_importances_df = pd.DataFrame({"feature": X_train.columns, "importance": rf.feature_importances_}).sort_values("importance", ascending = False)
feature_importances_df.head(20)
```

### Predicing on Test Data

```python
# Predicting on test data
y_pred = rf.predict(X_test)
```

```python
y_pred
```

```python
# Creating a dataframe to compare predicted and test values
pd.DataFrame({"y_pred": y_pred, "y_test": y_test})
```

### Calculating Mean Absolute Percentage Error (MAPE)

```python
# Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
mape
```

### XGBoost Regressor: Modelling the train data

```python
from xgboost import XGBRegressor
```

```python
# Training the Random Forest Regressor
xgb = XGBRegressor(n_estimators = 100, random_state = 0)
xgb.fit(X_train, y_train)
```

### Predicting on Test Data

```python
# Predicting on test data
y_pred = xgb.predict(X_test)
```

```python
y_pred
```

```python
# Creating a dataframe to compare predicted and test values
pd.DataFrame({"y_pred": y_pred, "y_test": y_test})
```

### Calculating Mean Absolute Percentage Error (MAPE)

```python
# Mean Absolute Percentage Error (MAPE)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
mape
```

#### When we drop Upvote and Downvote columns, fit our model on train data and try to test on test data, we see that MAPE increases considerably as compared when these two columns were included.
#### We got MAPE of 28.89 and 30 using Random Forest and XGBoost respectively excluding Upvote and Downvote. When Upvote and Downvote was included we got MAPE of 17.26 and 16.47 using Random Forest and XGBoost respectively.
